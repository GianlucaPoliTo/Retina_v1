```python
f = open("results_o.txt", "r")
string = f.read()
f.close()
```

```python
r = string.split("\n")
```

```python
rr = [x.split("?") for x in r if '' not in x.split("?")[0:12] ]
```

```python
import pandas as pd
name_col = rr.pop(0)
df = pd.DataFrame(rr, columns=name_col)
df.head()
```

```python
df['rtp.p_type']=df['rtp.p_type'].apply(lambda x: x.split(",")[0])
```

```python
df.head()
```

```python
import numpy as np
df = df.astype({'frame.time_epoch': 'float64', 'frame.number' : "int32", 'frame.len' : "int32", \
                'udp.length' : "int32", 'rtp.p_type' : "int32", \
                'rtp.timestamp' : np.int64, "udp.srcport" : "int32", \
                 "udp.dstport" : "int32", 'rtp.marker' : "int32", 'rtp.seq' : "int32", \
           })

df
```

```python
df = df.rename(columns ={
            'frame.time_epoch' : 'timestamps',
            'frame.number' : 'frame_num',
            'frame.len' : "len_frame",
            'ip.src' : 'ip_src',
            'ip.dst' : 'ip_dst',
            'udp.srcport' : 'prt_src',
            'udp.dstport' : 'prt_dst',
            'udp.length' :  'len_udp',
            'rtp.p_type' : 'p_type',
            'rtp.ssrc' : 'ssrc',
            'rtp.timestamp' : 'rtp_timestamp',
            'rtp.seq' : 'rtp_seq_num',
            'rtp.marker' : 'rtp_marker',
             'rtp.csrc.item' : 'rtp_csrc'
                })
df
```

```python
df["rtp_csrc"].replace('',"fec", inplace=True)
df
```

```python
columns = ["ssrc", "ip_src", "ip_dst", "prt_src", "prt_dst" , "p_type"]
gb= df.groupby(columns)
dict_flow_data = {x : gb.get_group(x) for x in gb.groups if x is not None and np.max(gb.get_group(x)["timestamps"]) - np.min(gb.get_group(x)["timestamps"])>2}
```
